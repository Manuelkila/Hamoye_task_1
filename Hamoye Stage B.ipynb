**Linear Regression Model for Living Room Temperature and Temperature outside Building** 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
linear_df = df[['Living_room_temp','tem_outside_building' ]]
linear_df.head(2)
```

```python
norm_linear_df = pd.DataFrame(scaler.fit_transform(linear_df), columns = linear_df.columns)
```

```python
norm_features =  norm_linear_df.drop(columns = ['tem_outside_building'])
norm_target = norm_linear_df['tem_outside_building']
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(norm_features, norm_target, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train1, y_train1)
```

```python
predicted_v = linear_model.predict(X_test1)
```

```python
from sklearn.metrics import mean_absolute_error
mae_o = mean_absolute_error(y_test1, predicted_v)
round(mae_o,3)
```

```python
rss_o = np.sum(np.square(y_test1 - predicted_v))
round(rss_o,3)
```

```python
from sklearn.metrics import mean_squared_error
rmse_o = np.sqrt(mean_squared_error(y_test1,predicted_v))
round(rmse_o,3)
```

```python
from sklearn.metrics import r2_score
r2_score_o = r2_score(y_test1, predicted_v)
round(r2_score_o,3)
```

**Multiple Linear Regression Model For the Dataset**

```python
df = df.drop(columns = ['date','lights'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
norm_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
features_df = norm_df.drop(columns = ['Appliances'])
target = norm_df['Appliances']
```

```python
features_df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_df, target, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
predicted_values = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae,3)
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss,3)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score,3)
```

**Ridge Model For The Dataset**

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(X_train, y_train)
```

```python
predicted_values2 =  ridge_reg.predict(X_test)
```

```python
mae2 = mean_absolute_error(y_test, predicted_values2)
round(mae,3)
```

```python
rss2 = np.sum(np.square(y_test - predicted_values2))
round(rss,2)
```

```python
rmse2 = np.sqrt(mean_squared_error(y_test,predicted_values2))
round(rmse,3)
```

**Lasso Model For the Dataset**

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha =0.001 )
lasso_reg.fit(X_train, y_train)
```

```python
predicted_values3 =  lasso_reg.predict(X_test)
```

```python
mae3 = mean_absolute_error(y_test, predicted_values3)
round(mae,3)
```

```python
rmse3 = np.sqrt(mean_squared_error(y_test,predicted_values3))
round(rmse,3)
```

**Model Weights**

```python
def get_weights_df(model,feat,col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, X_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, X_train, 'ridge_weight')
lasso_weights_df = get_weights_df(lasso_reg, X_train, 'lasso weight')
```

```python
final_weights = pd.merge(linear_model_weights,ridge_weights_df, on ='features')
final_weights = pd.merge(final_weights,lasso_weights_df, on = 'features')
```
